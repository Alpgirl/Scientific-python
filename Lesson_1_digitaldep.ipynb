hh.ru https://api.hh.ru Скачать по Москве 2000 вакансий по названию "Разработчик", построить гистограмму распределения зп по этим ваксниям, зп очистить от налогов и перевести зп из валютных единиц в рублевые.

```python
import requests
import numpy as np
import matplotlib.pyplot as plt
```

```python
#@title Текст заголовка по умолчанию
salary = []
# API выдает только 100 вакансий за раз, то есть максимум мы можем поднять 2000 вакансий
for i in range(0,20,1):
  url = "https://api.hh.ru/vacancies?text=Разработчик&area=1&search_field=name&per_page=100&page={}".format(str(i))
  items = requests.get(url).json()['items']
  for item in items:
    salary += [item['salary']]
```

```python
salary_clean = []
for sal in salary:
  if sal is not None:
    if sal['currency'] == "RUR":
      coef = 1
    else: coef = 0

    if sal['gross'] == True:
      gross = 1
    else: gross = 0.87

    if (sal['from'] is not None) and (sal['to'] is not None):
      salary_clean += [np.mean([sal['from'], sal['to']])*coef*gross]
    elif (sal['from'] is not None):
      salary_clean += [sal['from']*coef*gross]
    else: 
      salary_clean += [sal['to']*coef*gross]
salary_clean

#print(sal)
```

```python
plt.hist(salary_clean, bins=20)
```

```python
names = []
for i in range(0,20,1):
  url = "https://api.hh.ru/vacancies?text=Разработчик&area=1&search_field=name&per_page=100&page={}".format(str(i))
  items = requests.get(url).json()['items']
  for item in items:
    names += [item['name']]
```
